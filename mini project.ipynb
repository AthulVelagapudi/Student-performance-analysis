## Understanding test scores 


```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math score',data=sp,color='orange',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading score',data=sp,color='white',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing score',data=sp,color='green',linewidth=3)
plt.show()
```

From the vilonplot above we can see that the math,reading and writing scores are mostly in the range 50-80,45-75 and 40-80 respectively

## Relation between attributes and scores


```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='test preparation course',y='math score',data=sp)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='test preparation course',y='reading score',data=sp)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='test preparation course',y='writing score',data=sp)
plt.title('WRITING SCORES')
plt.show()
```

As we can see from the above barplots the students who have completed the test preparation have scored good than the students who have not completed the test preparation

```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(22,15))
plt.subplot(2, 2, 1)
sns.barplot(x='parental level of education',y='math score',data=sp)
plt.title('MATH SCORES')
plt.subplot(2, 2, 2)
sns.barplot(x='parental level of education',y='reading score',data=sp)
plt.title('READING SCORES')
plt.subplot(2, 2, 3)
sns.barplot(x='parental level of education',y='writing score',data=sp)
plt.title('WRITING SCORES')
plt.show()
```

From the above grpahs its clear that parental level of education changes the scores of the student there is a similar score for students whose parental level is high school and some high school and the rest of them are similar

```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='lunch',y='math score',data=sp)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='lunch',y='reading score',data=sp)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='lunch',y='writing score',data=sp)
plt.title('WRITING SCORES')
plt.show()
```

As we can see from the above barplots the students who have standard lunch tend to do better in their exams

```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='race/ethnicity',y='math score',data=sp)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='race/ethnicity',y='reading score',data=sp)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='race/ethnicity',y='writing score',data=sp)
plt.title('WRITING SCORES')
plt.show()
```

As we can see group E does better than other groups and group A does not perfer well as the other gorups

```python
plt.rcParams['axes.facecolor'] = "#b3ffff"
plt.rcParams['figure.facecolor'] ="#b3ffff"
plt.figure(figsize=(14,8))
plt.subplot(1, 3, 1)
sns.barplot(x='gender',y='math score',data=sp)
plt.title('MATH SCORES')
plt.subplot(1, 3, 2)
sns.barplot(x='gender',y='reading score',data=sp)
plt.title('READING SCORES')
plt.subplot(1, 3, 3)
sns.barplot(x='gender',y='writing score',data=sp)
plt.title('WRITING SCORES')
plt.show()
```

Males have better maths score compared to female where as female have better reading and writing scores than males
